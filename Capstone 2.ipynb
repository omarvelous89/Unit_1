# Research Proposal

## Questions
* Has Herd Immunity been compromised in California?
* Are Private schools more culpable for higher rates of exemption, i.e. lower rates of vaccination.
* Does the economic class of the child affect exemption rates?
* Do Medical Exemptions or Personal Belief Exemptions contribute more to the exemption rates in California?

## Hypothesis
* We will see increased rates of non-vaccination that are below threshold immunity.
* Null Hypothesis: non-vaccination rates will remain stable during the length of the study

* Private school will show a higher rate of non vaccination than Public schools.
* Null Hypothesis: Private schools will not show a higher rate of non-vaccination than Public Schools

* Economic class is a factor and families with higher incomes will also show a higher rate of non-vaccination.
* Null Hypothesis: There will be no difference in exemption rates based on economic status.

* Medical Exemptions contribute more to the overall exemption rate in California.
* Null Hypothesis: Personal Belief Exemptions contribute more to the overall exemption rate in California. 

## Participants
The participants in this study, or the population under analysis, is Kindergarten aged Californian students. The information was gathered by the schools the children attended and aggregated by the Counties/State. The students are only observed once per year. The data from each year represents a new population of the same school aged children.(Thus this is a between-subject design.

### Dependent Variable
Vaccination rate is my key metric. It is the measurable human behavior I evaluate my hypothesis by. My hypothesis states that the vaccination rate is dependent on school type, economic status, and the concept of herd immunity. 

Two other dependent variables are the Medical exemptions and the Personal Belief Exemptions. The raw counts for these variables are provided per school. The rates are calculated from the raw counts provided in the data divided by the class size.

These are good key metrics because they are also directly related. The overall vaccination rate and exemption rates should "complete" each other. They are approaching the questions from two different angles, Those who do vaccinate and those who do not

### Confounding variables
grade level is is a confounding variable due to the Law. However, for this study we have made grade level constant since we're running a between-subject experiment. 

Other confounding variables are the timing of the data being collected. Sometimes students may be un-vaccinated due to immigration status, or issues with their medical insurance but they present proof to the school of their intent to vaccinate. However, the school will report that the student is unvaccinated. We assume that these events are uncommon since vaccines are widely available and should be covered either by the State or 3rd party insurance.Additionally the chances that a child is missing a dose of a certain vaccine is assumed to be negligble.

#### Limitations of the Control group 
According to Mackenzie (p.167) "control variables compromise external validity by limiting responses to specific situations and people." This is true for this experiment as well. This analysis only validates my findings for a very specific age/grade level. It isn't generalizeable to the population of California.

However as we noted above, any sort of analysis across a less specific group (say including all grades k-12) would run the risk of Simpson's paradox. If I were presented with the k-12 data, I would still end up splitting the data into groups based on age because the law defines hard limits that should adhered to.

Laws are also not static. If there has been a change in the laws then there will almost certainly be a noticeable difference in the data. The point is that while this is a limitation of my control and the experiment as a whole, it's only a part of a larger analysis which accepts this limitation and works around it. 


## The method of testing the solution

* To test hypothesis one we compare the vaccination rates reported by the schools to the recommended rates we should see (according to the CDC and calculated through the threshold value formula.) So we can directly compare an observed value to a known value

* An alternative is also to compare the California population to other states in the country

* To test the second hypothesis we separate the data based on the public private label and then examine the data to see if the same vaccination rates we observed in the first part are present in the sliced data. We also control for the oversampling of the Public schools in California. Additionally we check to see if there have been any trends in the vaccination rates over time between the two groups.

* Using data obtained from the U.S. Census Bureau, we compare the the vaccination rate across different counties and group them based on median houshold income. We then compare those vaccination rates to the ideal ones we expect to see based on the first part of the experiment.
