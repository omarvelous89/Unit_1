```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

WHO = pd.read_csv('who_suicide_statistics.csv')
EGY = WHO.loc[lambda df:WHO['country'] == 'Egypt']
#EGY.groupby(['sex']).aggregate(np.mean)
EGY




```

```python
x = EGY.groupby(['sex', 'year'])
x.aggregate(np.mean)
```

```python
plt.figure(figsize=(10,7))
ax = sns.lineplot(x="year", y="suicides_no", hue="sex", data=EGY)

plt.title('Avg suicide per year per sex in Egypt', Fontsize=16);
plt.xlabel('Year', Fontsize=14);
plt.ylabel('Avg number of suicides', Fontsize=14);
```

```python
plt.figure(figsize=(15,7))
ax = sns.boxplot(x="year", y="suicides_no", hue="sex", data = EGY, palette="Set3")

plt.title('Suicides in Egypt', FontSize=16);
```

```python
plt.figure(figsize=(8,30))
plt.title('Suicide distribution in Egypt by sex');
sns.violinplot(y="year", x="suicides_no", hue="sex",
               split=True, inner="quart", orient='h' ,
               data=EGY)
sns.despine(left=True)

```

```python
plt.figure(figsize=(25,10));
g = sns.catplot(x="year", y="suicides_no", hue="age", col="sex", data=EGY, kind="swarm");
plt.xticks(rotation=45);

```

```python
plt.figure(figsize=(8,20));
plt.title('Suicides per year by sex and age range')
g = sns.FacetGrid(EGY, col="sex", row="age")
g = g.map(plt.plot, "year" , "suicides_no");
```

```python
plt.figure(figsize=(10,6));
plt.xticks(rotation=45)
ff = sns.countplot(x="age", hue="sex", data=EGY)
```
