```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

AP = pd.read_csv('appendix.csv')
AP
```

```python
AP = AP.rename(columns= {'Participants (Course Content Accessed)':'students' , '% Male':'male' , '% Female':'female' })
AP['nF'] = AP.students * (AP.female/100)
AP['nM'] = AP.students * (AP.male/100)
AP
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x=AP['Median Age'], orient = "h")
plt.title('Median Age of students');
```

```python
plt.hist(AP['Median Age'], bins=30)
plt.title('Median Age of Students')
plt.xlabel('Median Age');
```

```python
sns.stripplot(x=AP['Median Age'], y=AP['Institution'])
plt.title('Median Age and Institution');

```

```python
A = AP.loc[lambda df:AP['Course Number'] == '6.00x']
B = AP.loc[lambda df:AP['Course Number'] == 'CS50x']

frames = [A,B]

AB= pd.concat(frames)
AB

sns.violinplot(x=AB['Course Number'], y=AB['Median Age'], hue=AB['Institution']);
plt.title('Median age per Course')

```

```python
x1 = AB['Launch Date']
y1 = AB['nM']
y2 = AB['nF']

plt.figure(figsize=(10,8))

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('Female and Male student enrollment over time')
plt.ylabel('number of students')

plt.subplot(2, 1, 2)
plt.plot(x1, y2, '.-')
plt.xlabel('Launch Date');
plt.ylabel('number of students');

```

```python
d = AP.loc[lambda df:AP['students'] > 50000]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Distribution of female and male students', fontsize=16)
axs[0].set_ylabel('number of times')

axs[0].hist(AP['nF'], bins=10);
axs[0].set_xlabel('number of female students');


axs[1].hist(AP['nM'], bins=10);
axs[1].set_xlabel('number of male students');
```

```python
plt.figure(figsize=(12, 7))
plt.subplot(1, 2, 1)
fig.suptitle('student enrollment by sex', fontsize=18)

ax = sns.boxplot(  y="nF", x= "Institution", data=d,  orient='v').set_title('student enrollment female')
plt.yscale('log')


plt.subplot(1,2,2)
ax = sns.boxplot(  y="nM", x= "Institution", data=d,  orient='v').set_title('student enrollment male')
plt.yscale('log')
```
