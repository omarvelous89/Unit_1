## Motivation for this study

In the past few years the anti-vaccine movement became national news and their views became well-known. The language they used framed the issue as the government violating the right of the individual to choose what medical treatment they or their family can obtain. This movement found its justification in medical papers that found correlations between vaccines and Autism. These papers found that the links between Autism and Vaccines were significant and traced the cause to a preservative in the vaccines called Thiomersal, which contains Mercury as part of its molecular structure.

Families found themselves armed with a practical scapegoat in Mercury, a known poisonous substance, and a small group of medical professionals who could back them up. People began acting on these beliefs and obtained exemptions to vaccinations for their children. These exemptions were Personal Belief Exemptions and could be found across the country. The purpose of the exemptions was to accommodate those whose religious beliefs forbade them from receiving certain vaccines. It wasn't seen as a threat to public health because it was assumed that the amount of people seeking these exemptions would be so low that the public health cost of offering the Exemption was minimal.

However, if the number of people within a population seeking exemptions gets too large, then the immunity of the population, or herd, becomes compromised. Unvaccinated people are more likely to become a disease vector because their immune system is less likely to overcome an infection it hasn't encountered before. The unvaccinated person who becomes exposed, will incubate the bacteria and may become sick. If the virus or bacteria mutates inside the host, they can infect other people with the new strain. This is problematic because mutations can make vaccines less effective.  Herd immunity relies on the fact that if enough people are vaccinated, disease outbreaks are prevented because infection a chain of infections is less likely to occur.

## Questions

### H1
To test hypothesis 1 we will make use of histograms. The histogram bins will represent the vaccination rate present in the classes. There will be three histograms, one for each vaccine cocktail that we have data for, MMR( Measles/Mups/Rubella), DTP (Diptheria/Tetanus/Pertussis) and Polio. 

Our data only contains raw counts per vaccine, per class. However the data also provides us with the class size so the first step before making the histograms is calculating the vaccination rates by dividing the number of vaccinations in the class by the class size.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

SD = pd.read_csv('SD.csv')
SD['MMR_rate'] = SD.nMMR / SD.n
SD['DTP_rate'] = SD.nDTP / SD.n
SD['Polio_rate'] = SD.nPolio/ SD.n
SD['PBE_rate'] = SD.nPBE/SD.n
SD['PME_rate'] = SD.nPME/SD.n #Added 5 new columns to my dataframe. rate of vaccination per row.


```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 6))
fig.suptitle('Distribution of Vaccination rates for Kindergartners', fontsize=16)
axs[0].set_ylabel('number of instances')

axs[0].hist(SD['MMR_rate'], bins=20)
axs[0].set_xlabel('MMR Vaccination Rates');

axs[1].hist(SD['DTP_rate'], bins=20)
axs[1].set_xlabel('DTP Vaccination Rates');

axs[2].hist(SD['Polio_rate'], bins=20);
axs[2].set_xlabel('Polio Vaccination Rates');
```

$1-(1/R_0)$. is the formula that the CDC uses to calculate the threshold level of vaccination necessary in the population.

$R_0$ is called the *Reproductive Number*. It is calculated empirically and is unique to every infectious agent. $R_0$ measures the number of infected cases that can arise from an initial infection. For most diseases like Measles (found in the MMR vaccine), $R_0 = 12$. Ideally for a distribution of vaccination rates, one would expect to see very low distributions everywhere except around 95%-100% In the histograms.

## H3 Methodology
To test the second hypothesis we separate the data based on the public private label in School Type. We will then make box plot for each label the data where each box plot will represent a year from 2000-2014. The first set of boxplots will make use of the raw counts of vaccinations

The second set of boxplots will also be split by Private and Public but will make use of the calculated rates of vaccination instead of the raw counts. 

Public schools account for a much larger portion of the population (approximately 81,000 public schools to 28,000 Private schools) and also tend to have larger class sizes. The use of the Personal Belief rates instead of raw counts mitigates the class size bias just as it did for the vaccination rates as well.

However we need to control for the bias towards the amount of Public schools vs Private schools. The best way to account for this bias is by underampling the Public School data



```python
SDpriv = SD.loc[lambda df:SD['schoolType'] == 'PRIVATE'] #new dataframe with only private schools
SDpub = SD.loc[lambda df:SD['schoolType'] == 'PUBLIC'] #new dataframe with only public schools

SDU = SDpub.sample(n=28005)#creating an undersampled dataframe from Public school data, without replacement.
frames = [SDU, SDpriv] #set a variable equal to my two dataframes 
SDsampled = pd.concat(frames) #merge both dataframes together for a full undersampled DF

x1= SDsampled['year']
y1= SDsampled['nPBE']

x2= SDpriv['year']
y2= SDpriv['nPBE']
#defining variables for my plots

plt.figure(figsize=(15,8)) #subplot for my two boxplots
plt.subplot(1, 2, 1)
ax = sns.boxplot(x=x2, y=y2, data=SDpriv).set_title('Distribution of Personal Belief Exemptions in private schools');
plt.yscale("log");
plt.xticks(rotation=45);

plt.subplot(1,2,2)
ax = sns.boxplot(x=x1, y=y1, data=SDsampled).set_title('Distribution of Personal Belief Exemptions in public schools');
plt.yscale("log");
plt.xticks(rotation=45);

```

```python
x1= SDsampled['year']
y3= SDsampled['PBE_rate']

x2= SDpriv['year']
y4= SDpriv['PBE_rate']

plt.figure(figsize=(15,8))
plt.subplot(1, 2, 1)
ax = sns.boxplot(x=x2, y=y4, data=SDpriv).set_title('Distribution of PBE rate in private schools')
plt.xticks(rotation=45);
plt.subplot(1,2,2)
ax = sns.boxplot(x=x1, y=y3, data=SDsampled).set_title('Distribution of PPBE rate in public schools')
plt.xticks(rotation=45);
```
